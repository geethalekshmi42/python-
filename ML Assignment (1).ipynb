```python
#IQR method for outlier removal
def remove_outliers(data,columns):
  data_filtered=data.copy()

  for column in columns:
     Q1 = data[column].quantile(0.25)
     Q3 = data[column].quantile(0.75)
     IQR = Q3 - Q1

     lower_bound = Q1 - 1.5 * IQR
     upper_bound = Q3 + 1.5* IQR

     data_filtered = data_filtered[(data_filtered[column] >= lower_bound) & (data_filtered[column] <= upper_bound)]

  return data_filtered
```

```python
new_data=remove_outliers(data2,['radius error','perimeter error','smoothness error','concavity error','fractal dimension error'])
```
