```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
x = data2[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population','AveOccup','Latitude']]
y = data2['Longitude']
```

```python
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
print("Mean Squared Error (MSE):", mean_squared_error(y_test, Y_predict))
```

```python
print("R-Squared Error (RSE):", r2_score(y_test, Y_predict))
```

```python
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, Y_predict))
```
